Combine into one dataframe and sort by latitude

```python
heatmap_df3 = pd.DataFrame()
heatmap_df3['Q trends'] = q_list
heatmap_df3['AM7year trends'] = df['AM7year trend 5%']
heatmap_df3['AM30year trends'] = df['AM30year trend 5%']
heatmap_df3['AM7winter trends'] = df['AM7winter trend 5%']
heatmap_df3['AM30winter trends'] = df['AM30winter trend 5%']
heatmap_df3['AM7spring trends'] = spring7_list
heatmap_df3['AM30spring trends'] = spring30_list
heatmap_df3['AM7summer trends'] = df['AM7summer trend 5%']
heatmap_df3['AM30summer trends'] = df['AM30summer trend 5%']
heatmap_df3['AM7autumn trends'] = autumn7_list
heatmap_df3['AM30autumn trends'] = autumn30_list
heatmap_df3['Stations'] = df['Stations']
heatmap_df3['Lat'] = df['Lat']
heatmap_df3 = heatmap_df3.sort_values(by='Lat', ascending=False)
heatmap_df3=heatmap_df3.drop('Lat', axis=1)
heatmap_df3.index=heatmap_df3['Stations']
heatmap_df3=heatmap_df3.drop('Stations', axis=1)
```

Plot dataframe as heatmap

```python
fig, ax = plt.subplots(figsize=(15, 25))
sb.heatmap(heatmap_df3, cmap='RdBu', linewidths=0.5, annot=False)
plt.savefig('heatmap.png', transparent=False, facecolor='white')
```
